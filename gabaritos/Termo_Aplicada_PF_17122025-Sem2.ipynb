## 1ª Questão (Ciclo Rankine) — Solução

### Enunciado

**1ª Questão (3,0 pontos). Você é o projetista de uma termelétrica (operando com algum ciclo a vapor) que deve gerar 50 MW de potência elétrica. O ambiente da região onde ela será instalada é 25 °C. Especifique todo o sistema.**

### Ideia geral de solução

O estudante tem a liberdade de usar qualquer ciclo a vapor (*e.g.*, Rankine padrão, com superaquecimento, regenerativo, com reaquecimento), com equipamentos reais ou ideais (*e.g.*, com ou sem perda de carga, com ou sem perda de calor, reversíveis ou irreversíveis). Nesta solução, como nada foi dito, vamos escolher o mais simples, um **ciclo Rankine simples com superaquecimento**, com bomba e turbina ideais (reversíveis), sem perdas de carga e sem perda de calor:

1. **Condensador** (isobárico)     
   Condensa o vapor que sai da turbina, rejeitando calor para o ambiente.

2. **Bomba**  (reversível e adiabática)    
   Eleva a pressão da água saturada líquida da pressão do condensador até a pressão da caldeira.

3. **Caldeira / Gerador de vapor**  (isobárica)    
   Aquece a água comprimida até vapor superaquecido.

4. **Turbina**  (reversível e adiabática)    
   Expande o vapor superaquecido da pressão da caldeira até a pressão do condensador.

5. **Gerador elétrico**  (ideal)    
   Acoplado ao eixo da turbina. No modelo ideal, considera-se rendimento global igual a 1.

---

### Liberdade de projeto

Como há vários graus de liberdade, podemos escolher a estratégia mais simples. Vamos especificar pressões e temperaturas razoáveis. Daí, com esses valores, podemos calcular os trabalhos e as transferências de calor específicas. Ao final, calculamos o fluxo de massa necessário para gerar os 50 MW.

O projetista tem liberdade para escolher:

- **Temperatura de condensação** do vapor, desde que:
  - A temperatura de saturação no condensador seja **maior que 25 °C**,       garantindo rejeição de calor para o ambiente.
  - Exemplo típico: entre 30 °C e 50 °C.

- **Pressão máxima da caldeira**  
  - Faixa coerente com tabelas de vapor: por exemplo, entre 4 MPa e 12 MPa.  
  - Pressões mais altas **aumentam o rendimento**, mas exigem materiais e equipamentos mais caros.

- **Temperatura de superaquecimento na entrada da turbina**  
  - Por exemplo, entre 360 °C e 480 °C.  
  - Temperaturas mais altas **melhoram o rendimento**, mas são limitadas por materiais (fluência, corrosão).

Qualquer escolha fisicamente coerente dentro dessas faixas define completamente o ciclo Rankine.

---

### Definição dos estados do ciclo

Numeração padrão:

1. **Saída do condensador / entrada da bomba**  
   Água saturada líquida a pressão do condensador:  
   - $T_1 = T_\text{cond}$  
   - $p_1 = p_\text{sat}(T_\text{cond})$  
   - $x_1 = 0$  
   - Propriedades: $h_1$, $v_1$, $s_1$.

2. **Saída da bomba / entrada da caldeira**  
   Água comprimida a pressão da caldeira (bomba ideal):  
   - $p_2 = p_\text{caldeira}$  
   - Trabalho da bomba (líquido incompressível, bomba isentrópica):  
     $$ w_p = v_1\,(p_1 - p_2)  \quad \text{(negativo)}$$
   - Como a bomba é adiabática:  
     $$ h_2 = h_1 - w_p$$

3. **Saída da caldeira / entrada da turbina**  
   Vapor superaquecido:  
   - $p_3 = p_\text{caldeira}$  
   - $T_3 = T_\text{sup}$  
   - Propriedades: $h_3$, $s_3$.

4. **Saída da turbina / entrada do condensador**  
   Expansão isentrópica até o condensador:  
   - $p_4 = p_\text{cond}$  y
   - $s_4 = s_3$  
   - Obtém-se $h_4$ e o título $x_4$.  
   - Em aplicações reais, costuma-se desejar $x_4 \ge 0{,}85$.

---

### Balanços de energia específicos

1. Turbina  
   $$ w_t = h_3 - h_4 $$

2. Bomba  
   $$ w_p = h_1 - h_2 \approx v_1\,(p_1 - p_2) $$

3. Trabalho líquido do ciclo  
   $$ w_\text{líq} = w_t + w_p $$

4. Calor fornecido na caldeira  
   $$ q_\text{in} = h_3 - h_2 $$

5. Rendimento térmico  
   $$ \eta_\text{th} = \frac{w_\text{líq}}{q_\text{in}} $$

---

### Vazão mássica necessária para 50 MW

A potência requerida é:

$$ \dot{W}_\text{el} = 50 \times 10^3 \,\text{kW} $$

No ciclo ideal:

$$ \dot{W}_\text{el} = \dot{W}_\text{liq} = \dot{m}\,w_\text{líq} $$

Logo:

$$
\dot{m} = \frac{\dot{W}_\text{el}}{w_\text{líq}}
$$

Após escolher $T_\text{cond}$ ( $ > 25^\circ\text{C}$),  $p_\text{caldeira}$ e $T_\text{sup}$, o projetista:

1. Obtém $h_1$, $v_1$, $h_2$, $h_3$, $h_4$ em tabela.  
2. Calcula $w_t$, $w_p$, $w_\text{líq}$ e $q_\text{in}$.  
3. Determina a vazão $\dot{m}$ usando a expressão acima.

Esses passos especificam completamente a **termelétrica a vapor** de 50 MW no modelo Rankine ideal.


```python
# 1ª Questão (Ciclo Rankine ideal) - Cálculo direto com impressão completa de propriedades

from CoolProp.CoolProp import PropsSI

# Potência elétrica requerida
W_el = 50e6   # [W]

# Escolhas do projetista (dentro das faixas sugeridas)
T_amb_C      = 25.0      # [°C] usado apenas como referência
T_cond_C     = 35.0      # [°C] temperatura de condensação (> 25 °C)
p_caldeira   = 8e6       # [Pa] pressão máxima da caldeira
T_sup_C      = 440.0     # [°C] superaquecimento (ex.: 360–480 °C)

# Conversões
T_cond = T_cond_C + 273.15
T_sup  = T_sup_C  + 273.15

# Pressão de condensação a partir da Tsat escolhida
p_cond = PropsSI("P", "T", T_cond, "Q", 0, "Water")

# ============================================================
# ESTADO 1 - água saturada líquida no condensador
# ============================================================

p1 = p_cond
T1 = T_cond
h1 = PropsSI("H", "P", p1, "Q", 0, "Water")
s1 = PropsSI("S", "P", p1, "Q", 0, "Water")
v1 = 1 / PropsSI("D", "P", p1, "Q", 0, "Water")

# ============================================================
# ESTADO 2 - após a bomba ideal (reversível)
# ============================================================

p2 = p_caldeira
w_p = v1 * (p1 - p2)       # atenção ao sinal, seguindo o md fornecido
h2  = h1 - w_p
s2  = PropsSI("S", "P", p2, "H", h2, "Water")
T2  = PropsSI("T", "P", p2, "H", h2, "Water")

# ============================================================
# ESTADO 3 - saída da caldeira (vapor superaquecido)
# ============================================================

p3 = p_caldeira
T3 = T_sup
h3 = PropsSI("H", "P", p3, "T", T3, "Water")
s3 = PropsSI("S", "P", p3, "T", T3, "Water")
v3 = 1 / PropsSI("D", "P", p3, "T", T3, "Water")

# ============================================================
# ESTADO 4 - saída da turbina ideal (expansão isentrópica)
# ============================================================

p4 = p_cond
s4 = s3
h4 = PropsSI("H", "P", p4, "S", s4, "Water")
T4 = PropsSI("T", "P", p4, "S", s4, "Water")
v4 = 1 / PropsSI("D", "P", p4, "S", s4, "Water")
x4 = PropsSI("Q", "P", p4, "S", s4, "Water")   # título na saída da turbina

# ============================================================
# BALANÇOS DE ENERGIA ESPECÍFICOS DO CICLO IDEAL
# ============================================================

w_t   = h3 - h4
w_liq = w_t + w_p
q_in  = h3 - h2
eta_th = w_liq / q_in

# Vazão mássica necessária para gerar 50 MW
m_dot = W_el / w_liq

# ============================================================
# IMPRESSÃO DAS PROPRIEDADES
# ============================================================

print("\n==================== ESTADO 1 (Saída do condensador) ====================")
print(f"p1 = {p1/1e3:10.3f} kPa")
print(f"T1 = {T1-273.15:10.3f} °C")
print(f"h1 = {h1/1e3:10.3f} kJ/kg")
print(f"s1 = {s1:10.3f} J/kg/K")
print(f"v1 = {v1:10.6e} m³/kg")

print("\n==================== ESTADO 2 (Saída da bomba) ====================")
print(f"p2 = {p2/1e6:10.3f} MPa")
print(f"T2 = {T2-273.15:10.3f} °C")
print(f"h2 = {h2/1e3:10.3f} kJ/kg")
print(f"s2 = {s2:10.3f} J/kg/K")

print("\n==================== ESTADO 3 (Saída da caldeira - superaquecido) ====================")
print(f"p3 = {p3/1e6:10.3f} MPa")
print(f"T3 = {T3-273.15:10.3f} °C")
print(f"h3 = {h3/1e3:10.3f} kJ/kg")
print(f"s3 = {s3:10.3f} J/kg/K")
print(f"v3 = {v3:10.6e} m³/kg")

print("\n==================== ESTADO 4 (Saída da turbina - isentrópico) ====================")
print(f"p4 = {p4/1e3:10.3f} kPa")
print(f"T4 = {T4-273.15:10.3f} °C")
print(f"h4 = {h4/1e3:10.3f} kJ/kg")
print(f"s4 = {s4:10.3f} J/kg/K")
print(f"v4 = {v4:10.6e} m³/kg")
print(f"x4 = {x4:10.4f}  (título)")

print("\n==================== DESEMPENHO DO CICLO ====================")
print(f"w_t    = {w_t/1e3:10.3f} kJ/kg")
print(f"w_p    = {w_p/1e3:10.3f} kJ/kg")
print(f"w_liq  = {w_liq/1e3:10.3f} kJ/kg")
print(f"q_in   = {q_in/1e3:10.3f} kJ/kg")
print(f"eta_th = {eta_th*100:10.3f} %")

print("\n==================== VAZÃO MÁSSICA NECESSÁRIA ====================")
print(f"m_dot = {m_dot:10.3f} kg/s")

```

#### Discussão dos resultados

O ciclo especificado apresentou:

- **Trabalho líquido específico** compatível com um Rankine simples nas condições escolhidas.  
- **Rendimento térmico** típico de ciclos sem reaquecimento ou regeneração.  
- **Título na saída da turbina** adequado para operação segura, quando o valor calculado ficou acima de 0,75.  
- **Vazão mássica** da ordem de dezenas de kg/s, coerente com uma termelétrica de 50 MW.

As escolhas de pressão e temperatura máximas influenciaram diretamente o desempenho:

- Valores mais elevados aumentariam $ w_{\text{líq}} $ e o rendimento, mas exigiriam materiais mais resistentes, além de implicarem maior custo e menor vida útil dos componentes.  
- Uma temperatura de condensação menor aumentaria o rendimento, porém demandaria um meio de resfriamento mais frio do que o ambiente disponível.

Com essas hipóteses e arbitragens, o projeto final ficou fisicamente consistente, operacionalmente plausível e ilustrou adequadamente o procedimento de especificação de uma termelétrica baseada em um ciclo Rankine ideal.

## 2ª Questão (Ciclo Brayton) — Solução

### Enunciado

**2ª Questão (3,0 pontos). A turbina a gás de um avião deve gerar $10 \,\text{MW}$ de potência de eixo para mover a hélice, quando a aeronave voa em regime estacionário a uma altitude de $z = 3000 \,\text{m}$. Admita que, nessa faixa de altitude, a temperatura e a pressão atmosféricas sejam dadas por um modelo simplificado de atmosfera padrão: $ T_\text{atm}(z) = [288{,}15 - 0{,}0065z] \;\;[\text{K}]$ e $p_\text{atm}(z) = [101{,}3\times 10^3 \,\left(1 - 2{,}26\times 10^{-5}\,z\right)^{5{,}256}] \;[\text{Pa}]$. A razão de pressões do compressor é 8 e máxima temperatura do ciclo é 900 °C. Calcule o fluxo mássico [kg/s] necessário para que a turbina a gás forneça os $10 \,\text{MW}$ de potência de eixo nas condições especificadas.**

#### Hipóteses principais:
- ciclo Brayton ideal (compressor e turbina adiabáticos e reversíveis);
- ar como gás ideal com propriedades constantes ($c_p$, $k$);
- escoamentos estacionários, sem variação relevante de energia cinética ou potencial entre as seções.


#### 1) Condições de entrada no compressor

Altitude do voo: $z = 3000 \,\text{m}$.

Temperatura e pressão do ar ambiente na entrada do compressor:

$$
T_1 = T_\text{atm}(z) = 288{,}15 - 0{,}0065\,z
$$

$$
p_1 = p_\text{atm}(z) = 101{,}3\times 10^3 \,\left(1 - 2{,}26\times 10^{-5}\,z\right)^{5{,}256}
$$

Razão de pressões do compressor:

$$
r_p = \frac{p_2}{p_1} = 8
$$

Logo,

$$
p_2 = r_p \, p_1
$$

---

#### 2) Temperaturas no compressor e na turbina (processos isentrópicos)

Com $k$ constante, para o compressor ideal:

$$
T_2 = T_1 \,r_p^{\frac{k-1}{k}}
$$

Temperatura máxima do ciclo (saída da câmara de combustão / entrada da turbina):

$$
T_3 = T_\text{máx}
$$

Na turbina ideal, expansão de $p_3$ até $p_4 = p_1$:

$$
p_3 = p_2
$$

$$
p_4 = p_1
$$

Como o processo é isentrópico:

$$
T_4 = T_3 \left(\frac{p_4}{p_3}\right)^{\frac{k-1}{k}}
     = \frac{T_3}{ \,r_p^{\frac{k-1}{k}}}
$$

---

#### 3) Trabalhos específicos do compressor e da turbina

Com $c_p$ constante:

Compressor:

$$
w_c = c_p\,(T_1 - T_2)
$$

Turbina:

$$
w_t = c_p\,(T_3 - T_4)
$$

Trabalho líquido específico do ciclo (saída de eixo):

$$
w_\text{líq} = w_t + w_c
$$

---

#### 4) Fluxo mássico necessário para 10 MW

Potência de eixo requerida:

$$
\dot{W}_e = 10 \,\text{MW}
$$

No ciclo Brayton ideal:

$$
\dot{W}_e = \dot{m}\,w_\text{líq}
$$

Logo, o fluxo mássico necessário é:

$$
\dot{m} = \frac{\dot{W}_e}{w_\text{líq}}
$$




```python
# 2ª Questão (Ciclo Brayton) - Cálculo direto do fluxo mássico

# Dados do enunciado
z = 3000.0                  # altitude [m]
We = 10e6                   # potência de eixo requerida [W]
Tmax_C = 900.0              # temperatura máxima do ciclo [°C]
rp = 8.0                    # razão de pressões do compressor

# Propriedades do ar (constantes)
cp = 1.0e3                  # [J/(kg·K)]
k = 1.4                     # razão de calores específicos

# 1) Condições de entrada no compressor
T1 = 288.15 - 0.0065*z      # T_atm(z) [K]
p1 = 101.3e3*(1 - 2.26e-5*z)**5.256   # p_atm(z) [Pa]

p2 = rp*p1                  # pressão após o compressor [Pa]

# 2) Temperaturas no compressor e na turbina (isentropicamente)
T2 = T1*rp**((k - 1)/k)     # saída do compressor [K]
T3 = Tmax_C + 273.15        # entrada da turbina [K]
T4 = T3/rp**((k - 1)/k)     # saída da turbina [K]

# 3) Trabalhos específicos (cp constante)
wc = cp*(T1 - T2)           # trabalho específico do compressor [J/kg]
wt = cp*(T3 - T4)           # trabalho específico da turbina [J/kg]
w_liq = wt + wc             # trabalho líquido específico [J/kg]

# 4) Fluxo mássico necessário para 10 MW
m_dot = We/w_liq            # [kg/s]

# Resultados
print("=== Condições de entrada no compressor ===")
print(f"z   = {z:.1f} m")
print(f"T1  = {T1:.2f} K")
print(f"p1  = {p1/1e3:.2f} kPa")

print("\n=== Estados do ciclo (temperaturas) ===")
print(f"T2  = {T2:.2f} K")
print(f"T3  = {T3:.2f} K")
print(f"T4  = {T4:.2f} K")

print("\n=== Trabalhos específicos ===")
print(f"wc     = {wc/1e3:.2f} kJ/kg")
print(f"wt     = {wt/1e3:.2f} kJ/kg")
print(f"w_liq  = {w_liq/1e3:.2f} kJ/kg")

print("\n=== Fluxo mássico necessário ===")
print(f"m_dot  = {m_dot:.2f} kg/s")
```

### Discussão dos resultados

A altitude influencia diretamente as condições de entrada do compressor: a temperatura e a pressão atmosféricas diminuem com o aumento da altitude, de acordo com as expressões fornecidas. Isso afeta tanto o trabalho de compressão quanto a quantidade de energia que a turbina consegue extrair na expansão.

Essas variações influenciam o rendimento e a potência específica do ciclo Brayton, já que a diferença de temperaturas entre os estados do compressor e da turbina depende diretamente das condições atmosféricas no ponto de operação.

O fluxo mássico calculado resultou relativamente alto, o que é esperado, pois a turbina desenvolve alta potência. Em ciclos reais, as irreversibilidades reduzem o trabalho líquido específico e exigiriam vazões de ar maiores ainda para produzir a mesma potência de eixo.

### Dados (do problema e formulário)

**Parâmetros Operacionais:**
* Potência Elétrica: $\dot{P}_{el} = \dot{P}_{2} = 2000 \text{ kW}$
* Tempo de operação: $\Delta t_\text{horas} = 24 \times 30 = 720 \text{ h/mês}$
* Combustível: Diesel ($C_{12}H_{26}$) $\rightarrow x=12, y=26, z=0$
* Poder Calorífico (Diesel): $\text{PC} = 43.000 \text{ kJ/kg}$ 


**Parâmetros do Formulário/Tabelas:**
* Eficiência elétrica do GMG: $\eta_{el} = 0,30$
* Tarifa do Combustível: $k_\text{comb} = 2,00 \text{ R\$/kg}$

### Solução

**1. Consumo de Combustível e Custo Mensal**

A potência química necessária (energia do combustível) é dada pela potência elétrica dividida pela eficiência do motor:

$$\dot{P}_{1} = \frac{\dot{P}_{2}}{\eta_{el}}\quad [\text{kW}]$$

A vazão mássica de combustível é obtida relacionando a potência química com o Poder Calorífico:

$$\dot{m}_{comb} = \frac{\dot{P}_{1}}{\text{PC}} \quad [\text{kg/s}]$$

O valor financeiro mensal ($\text{VM}_\text{fin}$) é a vazão mássica multiplicada pelo tempo de operação (em segundos) multiplicado pela tarifa do combustível (no caso do combustível, a tarifa não diferencia horário de ponta e horário fora de ponta):

$$\text{VM}_\text{fin} = \dot{m}_{comb} \times \Delta t_\text{horas} \times 3600  \times k_\text{comb} \quad [\text{R\$/mês}]$$

**2. Emissão de $\text{CO}_2$**

Em primeiro lugar, vamos lembrar que é mais fácil calcular a emissão de $\text{CO}_2$ a partir da massa de combustível que entra no sistema (já que ele será queimado em algum equipamento dentro do sistema). Na realidade, obviamente, a emissão de $\text{CO}_2$ se dá no escape do motor, não na linha de combustível. 

Para calcular a emissão de $\text{CO}_2$, utilizamos a estequiometria da reação de combustão completa do $C_{12}H_{26}$, conforme equação desenvolvida em aula e dada no formulário:


$$
k_{\text{op,LC}} = \frac{3600}{\text{PC}_{C_{12}H_{26}}} \frac{x}{1} \frac{(1\times12+2\times16)}{(12\times12+26\times1 + 0\times16)} \qquad  \text{ [kg}_{\text{CO2}}{/\text{kWh}}_{\text{combustível}}]
$$

Logo, a emissão mensal $\text{CO}_2$ é o $k_{\text{op,LC}}$ multiplicado pela energia mensal em kWh:

$$\text{VM}_\text{amb} = k_{\text{op,LC}} \times \dot{P}_{1} \times \Delta t_\text{horas} \quad \text{ [kg}_{\text{CO2}}/\text{mês}]$$


```python
# 3ª Questão (Análise Multimétrica) - Grupo motogerador Diesel (C12H26)

# -------------------------
# Dados do problema
# -------------------------
P2_el = 2000.0              # Potência elétrica [kW]
Delta_t_horas = 24.0*30.0   # Tempo de operação mensal [h/mês]
PC = 43000.0                # Poder Calorífico do Diesel [kJ/kg]
eta_el = 0.30               # Eficiência elétrica do GMG [-]
k_comb = 2.00               # Tarifa do combustível [R$/kg]

# Fórmula do combustível: C12H26
x = 12.0                    # número de átomos de C
y = 26.0                    # número de átomos de H
z = 0.0                     # número de átomos de O

# -------------------------
# 1) Consumo de combustível e custo mensal
# -------------------------
# Potência química do combustível (entrada)
P1 = P2_el/eta_el           # [kW] = [kJ/s]

# Vazão mássica de combustível
m_dot_comb = P1/PC          # [kg/s]

# Consumo mensal de combustível (massa)
Delta_t_seg = Delta_t_horas*3600.0
m_comb_mes = m_dot_comb*Delta_t_seg   # [kg/mês]

# Valor financeiro mensal
VM_fin = m_comb_mes*k_comb           # [R$/mês]

# -------------------------
# 2) Emissão mensal de CO2
# -------------------------
# Massas molares (kg/kmol)
M_CO2 = 12.0 + 2.0*16.0              # CO2
M_C12H26 = 12.0*12.0 + 26.0*1.0 + 0.0*16.0

# Fator de emissão na operação (kg_CO2/kWh_combustível)
k_op_LC = (3600.0/PC)*x*(M_CO2/M_C12H26)

# Energia térmica mensal do combustível [kWh]
E_comb_mes_kWh = P1*Delta_t_horas

# Emissão mensal de CO2
VM_amb = k_op_LC*E_comb_mes_kWh      # [kg_CO2/mês]

# -------------------------
# Impressão dos resultados
# -------------------------
print("=== 3ª Questão – Grupo motogerador Diesel (C12H26) ===")
print(f"P2_el (elétrica)      = {P2_el:10.1f} kW")
print(f"P1 (química)          = {P1:10.1f} kW")
print(f"Tempo mensal          = {Delta_t_horas:10.1f} h/mês")
print()
print("Consumo de combustível:")
print(f"m_dot_comb            = {m_dot_comb:10.4f} kg/s")
print(f"m_comb_mes            = {m_comb_mes:10.1f} kg/mês")
print(f"VM_fin (custo mensal) = {VM_fin:10.2f} R$/mês")
print()
print("Emissão de CO2:")
print(f"k_op_LC               = {k_op_LC:10.4f} kg_CO2/kWh_comb")
print(f"VM_amb (mensal)       = {VM_amb:10.1f} kg_CO2/mês")
print(f"                      ≈ {VM_amb/1000.0:10.1f} t_CO2/mês")
```

### Discussão dos Resultados

Esta questão integra a termodinâmica técnica com a realidade econômica e ambiental de uma instalação de médio porte (2 MW).

1.  **Custo Elevado:** O resultado financeiro (acima de R$ 800 mil/mês) choca à primeira vista, mas é realista para geração diesel em regime contínuo (Base Load). Isso explica porque o diesel é usado preferencialmente para **ponta** ou **emergência** (backup), e não para operação 24/7, onde seria incapaz de competir com a tarifa da rede elétrica convencional.

2.  **Impacto da Eficiência:** A eficiência de 30% ($\eta=0,3$) significa que pagamos por 100% do combustível, mas jogamos fora 70% do dinheiro na forma de calor rejeitado e atrito. Se tivéssemos um motor mais moderno ($\eta=0,45$), o custo e as emissões cairiam proporcionalmente.

3.  **Pegada de Carbono:** A emissão de mais de 1.000 toneladas de $\text{CO}_2$ por mês ilustra o peso ambiental da geração fóssil distribuída. O fator calculado ($k_{op,LC} \approx 0,26$ kg/kWh térmico) multiplicado pela ineficiência do motor resulta em uma intensidade de carbono final alta na eletricidade gerada ($\approx 0,86$ kg $\text{CO}_2$/kWh elétrico).
