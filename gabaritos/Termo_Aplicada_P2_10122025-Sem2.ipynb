## 1ª Questão (Ciclo Diesel) — Solução

### Enunciado

**1ª Questão (4,0 pontos). Um motor Diesel de 2 litros, com taxa de compressão de 18:1 e rodando a 2500 rpm, aspira ar a 30°C e 100 kPa.  A combustão libera 800 kJ por kg de ar no cilindro.  Determine a máxima temperatura no ciclo [°C] e a potência mecânica do motor [kW].**

### Dados (enunciado)
- Volume deslocado total: $V_d$
- Taxa de compressão: $r_v = V_1/V_2$
- Rotação: $N_{\text{RPM}}$
- Condições de admissão: $p_{\text{amb}}$; $T_{\text{amb}}$
- Calor liberado durante a queima: $q_h=Q_h/m_{\text{cil}}$


### Dados (tabelas)

- Propriedades do ar: $c_{p0}$; $c_{v0}$; $R = c_{p0} - c_{v0}$; $k = c_{p0}/c_{v0}$

### Hipóteses
- Considerações comuns dos Ciclos Padrão a Ar (*i.e.*, o fluido de trabalho é sempre o ar, considerado gás perfeito com $cp_0$ constante; compressão e expansão isentrópicas; 
substituição da combustão e da exaustão/admissão por transferências de calor).
- No cálculo da massa do cilindro, iremos considerar que tanto os gases residuais quanto o ar aspirado estão a $T_{\text{amb}}$. Na verdade, os gases residuais deveriam estar a $T_4$. Já $T_1$ seria uma média ponderada entre $T_{\text{amb}}$ (resfriado pela evaporação do combustível) e $T_4$. O problema é que isso demandaria uma solução iterativa, já que $T_4$ depende de $T_1$, e $T_1$ por sua vez depende de $T_4$. Portanto, para simplificar, como dito no início, iremos considerar que no início da compressão a mistura está a $T_{\text{amb}}$

---

### 1. Cálculo das temperaturas do ciclo

#### 1.1 Compressão isentrópica $1 \to 2$

$$T_2 = T_1\, r_v^{\,k-1}$$

#### 1.2 Aquecimento a pressão constante $2 \to 3$



$$Q_h = m_{\text{cil}}c_{p0}(T_3 - T_2) \implies T_3 = T_2 + (Q_h/m_{\text{cil}})/c_{p0}= T_2 + q_h/c_{p0}$$  

$$T_{max} = T_3$$

#### 1.3 Expansão isentrópica $3 \to 4$

$$ p_3=p_2 \implies \dfrac{v_3}{T_3} = \dfrac{v_2}{T_2} \implies \dfrac{v_3}{v_2} = \dfrac{T_3}{T_2}$$  

$$v_4 = v_1$$

$$T_4 = T_3 \left( \dfrac{v_3}{v_4} \right)^{k-1} = T_3 \left( \dfrac{v_3/v_2}{v_1/v_2} \right)^{k-1} = T_3 \left( \dfrac{T_3/T_2}{r_v} \right)^{k-1}$$

---

### 2. Trabalho líquido específico

$$q_l = c_{v0}(T_1 - T_4) \quad \quad \text{(obs: }q_l \text{ será negativo})$$

$$w_{liq} = q_h + q_l \qquad{\text{[kJ por kg de ar no cilindro]}}$$


---

### 3. Massa por ciclo e vazão mássica

Como $\dot{W}_{liq} =  w_{liq} \times m_{\text{cil}}  \times \dot{n}_{\text{ciclos}}= $ (*i.e.*, a potência do motor é o trabalho específico, multiplicada pela massa no cilindro, multiplicada por quantos ciclos ocorrem por segundo), precisamos
agora calcular $m_{\text{cil}}$ e $\dot{n}_{\text{ciclos}}$.

Na Figura acima, vemos que a massa no cilindro durante o ciclo é dada pela soma da massa aspirada (*i.e.*, a massa aspirada enquanto o cilindro "desloca" um volume $V_d$ durante a admissão) com a massa residual do ciclo anterior (que ficou no volume $V_2$ ao final da descarga). Então, a relação entre $V_1$, $V_2$ e $V_d$ é dada por:

$$V_d = V_1 - V_2 = V_1(1 - 1/r_v) \quad \quad \text{(obs: } V_2 = V_1/r_v)$$

$$V_1 = \frac{V_d}{(1 - 1/r_v)}$$  

A massa no cilindro pode ser calculada por:

$$m_{\text{cil}} = m_{\text{d}} + m_{\text{res}} = \dfrac{p_{\text{amb}}\, V_d}{R\, T_{\text{amb}}} + \dfrac{p_{\text{amb}} \,V_2}{R\, T_{\text{4}}}$$

Neste momento, conforme dito na seção de hipóteses acima, vamos assumir que $p_4=p_1=p_{\text{amb}}$ e que $T_4=T_1=T_{\text{amb}}$ (para fugir de uma solução iterativa). Logo:

$$m_{\text{cil}} = \dfrac{p_1 \,V_d}{R \,T_1} + \dfrac{p_1 \,V_2}{R\, T_1}  = \dfrac{p_1\, V_1}{R \,T_1}  \qquad{\text{[kg de ar no cilindro durante o ciclo]}}$$

Observação: uma outra simplificação que o estudante poderia fazer, mais forte ainda , seria assumir que a massa do cilindro é apenas a massa do ar aspirado (*i.e.*, $m_{\text{cil}} = p_1 V_d/RT_1$). Com $r_v$=18, o volume residual é relativamente pequeno, 6%.

Lembrando que cada ciclo precisa de duas rotações:

$$\dot{n}_{ciclos} = N_{\text{RPM}}/(2\times 60) \qquad \text{[ciclos/segundo]}$$  

Temos, finalmente:

$$\dot{m} =  \dot{n}_{ciclos} \,m_{\text{cil}}= \frac{N_{\text{RPM}}}{2\times 60}\dfrac{p_1\, V_1}{R \,T_1}$$

---

### 4. Potência mecânica

Com o valor de $w_{liq}$ e $\dot{m}$ calculados na seção anterior, é fácil agora determinar a potência mecânica do motor:

$$\dot{W} = w_{liq}\, \dot{m}$$

---

### Resultados

- Temperatura máxima: $T_{max} = T_3$  
- Potência mecânica: $\dot{W} = w_{liq}\,\dot{m}$


```python
# ---------------------------------------------------------
# Cálculo do ciclo Diesel padrão a ar (versão simplificada)
# ---------------------------------------------------------

import numpy as np

# ----------------------------------
# Dados de entrada do problema
# ----------------------------------
Vd   = 2.0e-3        # volume deslocado total (m^3)
rv   = 18            # taxa de compressão = V1/V2
N    = 2500          # rotação (rpm)
p1   = 100e3         # pressão ambiente (Pa)
T1   = 30 + 273.15   # temperatura ambiente (K)
qh   = 800e3         # calor liberado por kg de ar (J/kg)

# Dados do ar
cp0  = 1.005e3       # J/(kg.K)
cv0  = 0.718e3       # J/(kg.K)
R    = cp0 - cv0
k    = cp0/cv0

# ----------------------------------------
# 1) Geometria: V1, V2 e massa no cilindro
# ----------------------------------------
V1 = Vd/(1 - 1/rv)   # volume total no início da compressão
V2 = V1/rv           # volume no final da compressão

# Massa no cilindro (hipótese simplificada: T_residual = Tamb)
m_cil = p1*V1/(R*T1)

# ----------------------------------------
# 2) Temperaturas do ciclo Diesel ideal
# ----------------------------------------
T2 = T1 * rv**(k-1)                    # Compressão isentrópica
T3 = T2 + qh/cp0                       # Aquecimento pcte
T4 = T3 * ((T3/T2)/rv)**(k-1)          # Expansão isentrópica

# -------------------------
# 3) Trabalho específico
# -------------------------
q_l   = cv0*(T1 - T4)       # rejeição de calor (negativa)
wliq  = qh + q_l            # trabalho líquido específico (J/kg)

# -------------------------
# 4) Vazão mássica e potência
# -------------------------
ndot_ciclos = N/(2*60)         # ciclos/s (motor 4 tempos)
mdot     = m_cil * ndot_ciclos
Wdot     = wliq * mdot      # potência (W)

# -------------------------------------------------
# 5) Impressão dos resultados de forma organizada
# -------------------------------------------------

print("=== TEMPERATURAS DO CICLO (K) ===")
print(f"T1 (início da compressão)   = {T1:.1f}")
print(f"T2 (fim da compressão)      = {T2:.1f}")
print(f"T3 (ponto de maior temp.)   = {T3:.1f}")
print(f"T4 (fim da expansão)        = {T4:.1f}")
print()

print("=== ENERGIAS ESPECÍFICAS (kJ/kg) ===")
print(f"q_h (adicionado)            = {qh/1e3:.1f}")
print(f"q_l (rejeitado)             = {q_l/1e3:.1f}")
print(f"w_liq (líquido)             = {wliq/1e3:.1f}")
print()

print("=== GEOMETRIA E MASSAS ===")
print(f"V1 (volume total)           = {V1:.6f} m³")
print(f"V2 (volume comprimido)      = {V2:.6f} m³")
print(f"m_cil (massa por ciclo)     = {m_cil:.5f} kg")
print(f"mdot (vazão mássica)        = {mdot:.5f} kg/s")
print()

print("=== POTÊNCIA MECÂNICA ===")
print(f"Wdot = {Wdot/1e3:.2f} kW")
print()

# -------------------------------------------------
# 6) RESULTADOS PEDIDOS NO ENUNCIADO
# -------------------------------------------------
print("=== RESULTADOS PEDIDOS NO ENUNCIADO ===")
print(f"Temperatura máxima (°C): {T3 - 273.15:.1f} °C")
print(f"Potência mecânica (kW):  {Wdot/1e3:.2f} kW")
```

### Discussão dos Resultados

Os resultados numéricos obtidos no código revelam a essência do funcionamento do ciclo Diesel:

O cálculo mostra que, apenas comprimindo o ar 18 vezes ($r_v=18$), atingimos uma temperatura muito superior à de autoignição do óleo diesel ($\approx 210\text{--}260^\circ\text{C}$). Isso prova matematicamente por que este motor dispensa velas de ignição: o temperatura da carga durante a injeção de combustível é suficiente para iniciar a combustão.

A temperatura máxima do ciclo é altíssima. Isso demonstra a necessidade de termos sistemas de arrefecimento (radiador/água) e lubrificação eficientes para evitar a fusão ou travamento dos componentes metálicos.

A temperatura de saída é muito maior que a de entrada. Isso indica que uma quantidade enorme de energia (entalpia) é jogada fora pelo escape, o que é uma indicação do relativo baixo rendimento do cicl.

**O que muda no Motor Real?**

O modelo calculado acima é o "Padrão a Ar Frio" (ideal). Um motor real apresentaria uma potência menor e um consumo maior devido a quatro fatores principais que ignoramos aqui:

1.  **Variação das Propriedades do Ar ($k$ variável):** No código, usamos $k=1,4$ fixo. Na realidade, conforme a temperatura sobe para 1500°C, o $k$ do ar diminui, o que reduz o rendiumento do ciclo. Esta é a principal fonte de erro teórico.
2.  **Perdas de Calor (Não-Adiabático):** O motor real perde calor pelas paredes do cilindro para a água de refrigeração. Isso "rouba" energia que seria usada para empurrar o pistão, diminuindo a temperatura $T_3$ e o trabalho final.
3.  **Combustão Imperfeita:** A queima do combustível nunca é instantânea nem completa. O atraso na queima e a formação de fuligem significam que nem todos os 800 kJ/kg são liberados efetivamente.
4.  **Atrito e Bombeamento:** O nosso modelo calcula a força no topo do pistão (*Potência Indicada*). Na prática, perdemos de 10% a 20% dessa energia vencendo o atrito dos anéis e sugando/empurrando os gases (*Potência Efetiva*).

## 2ª Questão (Emissão de $\text{CO}_{2}$) — Solução

### Enunciado

**2ª Questão (3,0 pontos)**. Uma fábrica necessita de 100 kW de energia elétrica, durante 8 horas por dia, apenas nos dias úteis. Há duas opções: [A] gerar energia elétrica com um grupo motogerador (queimando CH4, PCI de 45.000 kJ/kg) e [B] comprar energia elétrica diretamente da rede (a rede gera energia elétrica por uma termelétrica a carvão mineral). Calcule quantos kg de $\text{CO}_{2}$ cada um dos sistemas emite por mês, considerando apenas a operação (*i.e.*, você pode desprezar a emissão de $\text{CO}_{2}$ devido à construção dos equipamentos).

### Dados (enunciado)

- $x=1; y=4; z=0$
- $\text{PCI}_{C_{x}H_{y}O_{z}} = 45000 \text{ kJ/kg} \quad$ (poder calorífico do metano)
- $h_{op} = 8 \times 21 = 168\text{ h} \quad$ (horas mensais de operação) 
- $\dot{P}_{\text{2,A}} = \dot{P}_{\text{1,B}}=100 \text{ kW} \quad $  Potência elétrica demandada pela fábrica, nos dois sistemas.

### Dados (formulário)

- $k_{\text{op,UTE}} = 0,972 \text{ kg}_{\text{CO2}}{/\text{kWh}}_{\text{eletricidade}} \quad$  (emissão específica de $\text{CO}_{2}$ da termelétrica) 
- $\eta_{\text{el,GMG}} = 0,3 \quad$  (eficiência elétrica do grupo motogerador) 

### Solução

  <img src="../figures/Termo_Aplicada_P2_10122025-fig-b.png"
       alt=""
       width="420" style="display:block; margin:0 auto; max-width:100%;">
  <figcaption><b>Figura 2.</b> Sistemas de fornecimento elétrico.</figcaption>
</figure> 

Os dois sistemas são apresentados na figura acima. Desconsiderando a emissão de $\text{CO}_{2}$ durante a construção dos equipamentos, temos que:

$$
   \text{VM}_{\text{amb,A}} = k_{\text{op,LC}} \times \dot{P}_{\text{1,A}} \times h_{op}
$$

$$
   \text{VM}_{\text{amb,B}} = k_{\text{op,UTE}} \times \dot{P}_{\text{1,B}} \times h_{op}
$$


Devemos lembrar que, obviamente, a emissão de $\text{CO}_{2}$ se dá no motor, não na linha de combustível. Mas, como todo combustível que entra na planta será queimado em algum equipamento ($e.g.$, motor, caldeira), é mais fácil concentrar o cálculo todo na linha de combustível ao fazer os cálculos.


Sabemos que: 

$$\dot{P}_{\text{2,A}} = \eta_{\text{el,GMG}} \times \dot{P}_{\text{1,A}}  \implies \dot{P}_{\text{1,A}} = \frac{\dot{P}_{\text{2,A}}}{\eta_{\text{el,GMG}}}$$

e que:

$$
k_{\text{op,LC}} = \frac{3600}{\text{PCI}_{C_{x}H_{y}O_{z}}} \frac{x}{1} \frac{(1\times12+2\times16)}{(x\times12+y\times1 + z\times16)} \qquad  \text{ [kg}_{\text{CO2}}{/\text{kWh}}_{\text{combustível}}]
$$


Finalmente, chegamos a:


$$
   \text{VM}_{\text{amb,A}} = k_{\text{op,LC}} \times \frac{\dot{P}_{\text{2,A}}}{\eta_{\text{el,GMG}}} \times h_{op}
$$

$$
   \text{VM}_{\text{amb,B}} = k_{\text{op,UTE}} \times \dot{P}_{\text{1,B}} \times h_{op}
$$





```python
# -----------------------------------------------
# 2ª Questão – Emissão mensal de CO2 (Sistemas A e B)
# -----------------------------------------------

# ---- Dados do enunciado ----
P_elec = 100.0          # kW de eletricidade demandada pela fábrica
h_op   = 8*21           # h/mês (8 h/dia × 21 dias úteis)

# Combustível do GMG: CH4 ~ CxHyOz com x=1, y=4, z=0
x, y, z = 1, 4, 0
PCI = 45000.0           # kJ/kg (metano)

# ---- Dados do formulário ----
k_op_UTE   = 0.972      # kg_CO2 / kWh_electricidade (UTE a carvão)
eta_el_GMG = 0.30       # eficiência elétrica do grupo motogerador

# ---- 1) Fator de emissão do combustível (linha de combustível) ----
# M_CO2 = 44 kg/kmol; M_comb = 12x + y + 16z (kg/kmol)
M_CO2   = 44.0
M_comb  = 12.0*x + 1.0*y + 16.0*z

# 3600/PCI → kg_comb / kWh_comb
# (x*M_CO2/M_comb) → kg_CO2 / kg_comb
k_op_LC = (3600.0/PCI) * (x*M_CO2/M_comb)   # kg_CO2 / kWh_combustivel

# ---- 2) Sistema A: GMG a CH4 ----
P_quim_A = P_elec / eta_el_GMG             # kW de combustível na entrada do GMG
VM_amb_A = k_op_LC * P_quim_A * h_op       # kg_CO2 emitidos por mês (Sistema A)

# ---- 3) Sistema B: Rede elétrica (UTE a carvão) ----
P_elec_B = P_elec                          # mesma potência elétrica demandada
VM_amb_B = k_op_UTE * P_elec_B * h_op      # kg_CO2 emitidos por mês (Sistema B)

# Conversão para toneladas por mês
VM_amb_A_t = VM_amb_A / 1000.0
VM_amb_B_t = VM_amb_B / 1000.0

# ---- 4) Impressão organizada ----
print("=== FATORES DE EMISSÃO ===")
print(f"k_op_LC  (CH4, combustível) = {k_op_LC:.3f} kg CO2/kWh_comb")
print(f"k_op_UTE (carvão, eletric.) = {k_op_UTE:.3f} kg CO2/kWh_elec")
print()

print("=== EMISSÕES MENSAIS DE CO2 ===")
print(f"Sistema A (GMG a CH4) : {VM_amb_A:8.1f} kg CO2/mês  ({VM_amb_A_t:5.2f} t/mês)")
print(f"Sistema B (Rede/UTE)  : {VM_amb_B:8.1f} kg CO2/mês  ({VM_amb_B_t:5.2f} t/mês)")
print()

reducao_rel = (1 - VM_amb_A/VM_amb_B)*100
print("=== COMPARAÇÃO ===")
print(f"O Sistema A emite {reducao_rel:.1f}% menos CO2 que o Sistema B.")
```

### Discussão dos Resultados

A análise comparativa entre a geração local (Sistema A) e o consumo da rede (Sistema B) revela que a opção **[A] Geração Local a Gás Natural** é ambientalmente superior em termos de emissões de carbono, emitindo menos $\text{CO}_2$ por mês do que a opção da rede baseada em carvão.

#### Por que isso acontece?

O resultado depende do balanço entre **química do combustível** e **eficiência da conversão**:

1.  **Vantagem Química (Combustível):** O Metano ($CH_4$) possui uma alta proporção de Hidrogênio para Carbono ($H:C = 4:1$). Isso significa que grande parte da energia liberada vem da formação de água ($H_2O$), e não de $CO_2$. Já o carvão mineral é rico em carbono, gerando muito mais $CO_2$ por unidade de energia liberada.
2.  **Desvantagem da Eficiência:** O gerador local opera com uma eficiência baixa ($\eta = 30\%$), o que nos obriga a queimar muito mais combustível (entrada térmica de $333 \text{ kW}$ para gerar $100 \text{ kW}$ elétricos).

**Conclusão:** Neste caso específico, mesmo com a desvantagem da baixa eficiência do motor pequeno, a "limpeza" química do gás natural supera a alta intensidade de carbono do carvão da rede. O Sistema A apresenta uma intensidade de carbono global de aprox. **0,73 kg $\text{CO}_2$/kWh**, enquanto a rede opera a **0,97 kg $\text{CO}_2$/kWh**.

#### 2. Modelagem financeira

A figura acima mostra os dois sistemas propostos no enunciado. Em cada um deles, temos que:

$$\text{VP}_{\text{fin}} = \sum \text{VI}_{\text{i,fin}} + \mu \text{VM}_{\text{fin}} $$

onde $\mu$, número de meses corrigido, é calculado por:

$$ \mu = \frac{(1+j)^n - 1}{j(1+j)^n }  $$  

O valor inicial de cada equipamento é calculado por:

$$ \text{VI}_{\text{i,fin}} = k_{\text{inv,i}} \times \dot{P}_{\text{saída,i}} $$

E o valor mensal é calculado por (caso simplificado, onde as potências são constantes ao longo do tempo):


$$ \text{VM}_{\text{fin}} = k_{\text{cons,hp}}\times \dot{P}_{\text{RD}} \times h_{\text{op,hp}} + k_{\text{cons,fp}}\times \dot{P}_{\text{RD}} \times h_{\text{op,fp}} + k_{\text{dem,hp}}\times \dot{P}_{\text{RD}} + k_{\text{dem,fp}}\times \dot{P}_{\text{RD}} +  \underbrace{\frac{k_{\text{comb}}\times 3600}{\text{PCI}_{\text{comb}}}}_{\text{R\$/kg}\rightarrow \text{R\$/kWh}}   \times \dot{P}_{\text{LC}} \times (h_{\text{op,hp}} + h_{\text{op,fp}})$$


#### 3. Cálculo das potências

Agora, precisamos usar as eficiências e COPs dos equipamentos para calcular as potências. No sistema C:

$$\dot{P}_{\text{1,C}} = \frac{\dot{P}_{\text{2,C}}}{\text{COP}_{\text{CC}}}$$
$$\dot{P}_{\text{2,D}} = \frac{\dot{P}_{\text{3,D}}}{\text{COP}_{\text{CA}}}$$
$$\dot{P}_{\text{1,D}} = \frac{\dot{P}_{\text{2,D}}}{\eta_{\text{CQD}}}$$



#### 4. Equacionamento final

Substituindo as potências de cada sistema nas equações financeiras, chegamos ao seguinte valor presente para o sistema C:

$$\text{VP}_{C,fin} = \left\{ k_{\text{inv,RD}} \times \dot{P}_{\text{1,C}} + k_{\text{inv,CC}} \times \dot{P}_{\text{2,C}}\right\} + \mu \left\{ \xi (k_{\text{cons,hp}}\times \dot{P}_{\text{1,C}} \times h_{\text{op,hp}} + k_{\text{cons,fp}}\times \dot{P}_{\text{1,C}} \times h_{\text{op,fp}} + k_{\text{dem,hp}}\times \dot{P}_{\text{1,C}} + k_{\text{dem,fp}}\times \dot{P}_{\text{1,C}}) +  0 \right\} $$

ou

$$\text{VP}_{C,fin} = \text{VI}_{C},fin + \xi \mu \text{VM}_{C\text{,base,fin}}$$

onde:

$$
  \text{VI}_{C,fin} = k_{\text{inv,RD}} \times \dot{P}_{\text{1,C}} + k_{\text{inv,CC}} \times \dot{P}_{\text{2,C}}
$$

e

$$
\text{VM}_{C\text{,base,fin}} = k_{\text{cons,hp}}\times \dot{P}_{\text{1,C}} \times h_{\text{op,hp}} + k_{\text{cons,fp}}\times \dot{P}_{\text{1,C}} \times h_{\text{op,fp}} + k_{\text{dem,hp}}\times \dot{P}_{\text{1,C}} + k_{\text{dem,fp}}\times \dot{P}_{\text{1,C}} 
$$





Para o sistema D, temos o seguinte valor presente:

$$\text{VP}_{D,fin} = \left\{ k_{\text{inv,LC}} \times \dot{P}_{\text{1,D}} + k_{\text{inv,CQD}} \times \dot{P}_{\text{2,D}} + k_{\text{inv,CA}} \times \dot{P}_{\text{3,D}}   \right\} + \mu \left\{ 0 + 0 + 0 + 0 + \frac{k_{\text{comb}}\times 3600}{\text{PCI}_{\text{comb}}}   \times \dot{P}_{\text{1,D}} \times (h_{\text{op,hp}} + h_{\text{op,fp}})  \right\} $$

ou: 


$$\text{VP}_{D,fin} = \left\{ k_{\text{inv,LC}} \times \dot{P}_{\text{1,D}} + k_{\text{inv,CQD}} \times \dot{P}_{\text{2,D}} + k_{\text{inv,CA}} \times \dot{P}_{\text{3,D}}   \right\} + \mu \left\{ \frac{k_{\text{comb}}\times 3600}{\text{PCI}_{\text{comb}}}   \times \dot{P}_{\text{1,D}} \times (h_{\text{op,hp}} + h_{\text{op,fp}})  \right\} $$



Finalmente, igualando $\text{VP}_{C,fin} = \text{VP}_{D,fin}$, chegamos à:


$$ \xi = \frac{\text{VP}_{D,fin}-\text{VI}_{C,fin} }{\mu \text{VM}_{C\text{,base,fin}}}  $$





```python
# 3ª Questão – Análise financeira do chiller de absorção
# Objetivo: encontrar o fator ξ para o qual VP_C(ξ) = VP_D

# -----------------------------
# 1. Dados de entrada
# -----------------------------

# Potência de água gelada (ambos os sistemas)
P_AG = 200.0             # kW

# Horas de operação por mês
h_hp  = 3 * 21           # h em horário de ponta
h_fp  = 9 * 21 + 12 * 9  # h fora de ponta
h_tot = h_hp + h_fp      # h totais

# Dados financeiros
j = 0.01                 # taxa de juros mensal
n = 60                   # número de meses

k_inv_RD  = 100.0        # R$/kW_EE   (rede)
k_inv_CC  = 500.0        # R$/kW_AG   (chiller compressão)
k_inv_LC  = 100.0        # R$/kW_comb (linha de combustível)
k_inv_CQD = 300.0        # R$/kW_VP   (caldeira queima direta)
k_inv_CA  = 1500.0       # R$/kW_AG   (chiller absorção)

k_cons_hp = 0.7          # R$/kWh_EE  (consumo, ponta)
k_cons_fp = 0.2          # R$/kWh_EE  (consumo, fora de ponta)
k_dem_hp  = 70.0         # R$/kW_EE   (demanda, ponta)
k_dem_fp  = 20.0         # R$/kW_EE   (demanda, fora de ponta)

k_comb    = 2.0          # R$/kg_comb
PCI_comb  = 35000.0      # kJ/kg_comb

COP_CC = 4.0             # chiller compressão
COP_CA = 0.7             # chiller absorção
eta_CQD = 0.8            # eficiência caldeira

# -----------------------------
# 2. Potências nos sistemas
# -----------------------------

# Sistema C (compressão)
P2_C = P_AG                           # kW de água gelada
P1_C = P2_C / COP_CC                  # kW elétricos na rede

# Sistema D (absorção)
P3_D = P_AG                           # kW de água gelada
P2_D = P3_D / COP_CA                  # kW de calor no absorvedor
P1_D = P2_D / eta_CQD                 # kW de combustível na caldeira

# -----------------------------
# 3. Fator de valor presente μ
# -----------------------------

mu = ((1 + j)**n - 1) / (j * (1 + j)**n)

# -----------------------------
# 4. Investimento inicial
# -----------------------------

VI_C = k_inv_RD * P1_C + k_inv_CC * P2_C

VI_D = (
    k_inv_LC  * P1_D +
    k_inv_CQD * P2_D +
    k_inv_CA  * P3_D
)

# -----------------------------
# 5. Custos mensais (ξ = 1 na base)
# -----------------------------

# Sistema C – termo base que será multiplicado por ξ
VM_C_base = P1_C * (
    k_cons_hp * h_hp +
    k_cons_fp * h_fp +
    k_dem_hp  +
    k_dem_fp
)

# Sistema D – só combustível
c_comb_kWh = k_comb * 3600.0 / PCI_comb   # R$/kWh_comb
VM_D = c_comb_kWh * P1_D * h_tot          # R$/mês

# -----------------------------
# 6. Valor presente dos sistemas
# -----------------------------

# VP_C(ξ) = VI_C + μ * ξ * VM_C_base
# VP_D    = VI_D + μ * VM_D  (não depende de ξ)

VP_D = VI_D + mu * VM_D

# Fator crítico ξ* tal que VP_C(ξ*) = VP_D
xi_star = (VP_D - VI_C) / (mu * VM_C_base)

# -----------------------------
# 7. Impressão dos resultados
# -----------------------------

print("=== Potências ===")
print(f"P1_C (rede, sistema C)       = {P1_C:7.2f} kW")
print(f"P1_D (combustível, sist. D)  = {P1_D:7.2f} kW")
print()

print("=== Investimentos iniciais ===")
print(f"VI_C (compressão)            = {VI_C:10.2f} R$")
print(f"VI_D (absorção)              = {VI_D:10.2f} R$")
print()

print("=== Custos mensais base (ξ = 1) ===")
print(f"VM_C_base (sist. C)          = {VM_C_base:10.2f} R$/mês")
print(f"VM_D (sist. D)               = {VM_D:10.2f} R$/mês")
print()

print("=== Valor presente do sistema D ===")
print(f"VP_D                         = {VP_D:10.2f} R$")
print()

print("=== Fator crítico de tarifa elétrica ===")
print(f"xi* (tarifas elétricas)      = {xi_star:5.2f}")
print("Para ξ > xi*, o sistema com chiller de absorção (D) passa a ser mais econômico.")
```
